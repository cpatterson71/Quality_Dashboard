```python
import pandas as pd
import numpy as np
import altair as alt
import xlwings as xl
import plotly.express as px
```

```python
file = 'Quality_System_Document_Use_This.xlsx'
wb1 = xl.Book(file)
sht = wb1.sheets[0]
temp = sht.range('A1:N1322')
temp = temp.options(pd.DataFrame, index=False).value
temp.head()
```

```python
temp.columns
```

```python
temp = temp.loc[:, ['Date Opened', 'Change Control', 'Client', 'Requester', 'Document', 'Revision', 'Reviewer', 'Approver', 'Date Closed ', 'Days Open']]
temp.head()
```

```python
weeks = temp['Date Closed '].astype(str)
days = []
for week in weeks:
    days.append(week[:-9])
temp['Date Closed '] = days

temp.head()
```

```python
temp.dtypes
```

```python
temp['Date Opened'] = temp['Date Opened'].dt.strftime('%m/%d/%Y')
temp.head()
```

```python
temp.to_csv('temp.csv', index=False)
```
