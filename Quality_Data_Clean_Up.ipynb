```python
import pandas as pd
import numpy as np
import altair as alt
import xlwings as xl
import plotly.express as px
import datetime as dt
```

```python
file = 'Quality_System_Document_Use_This.xlsx'
wb1 = xl.Book(file)
sht = wb1.sheets[0]
temp = sht.range('A1:N1322')
temp = temp.options(pd.DataFrame, index=False).value
temp.head()
```

```python
temp = temp.loc[:, ['Date Opened', 'Change Control', 'Client', 'Requester', 'Document', 'Revision', 'Reviewer', 'Approver', 'Date Closed ', 'Days Open']]
temp.head()
```

```python
weeks = temp['Date Closed '].astype(str)
days = []
for week in weeks:
    days.append(week[:-9])
temp['Date Closed '] = days

temp.head()
```

```python
temp['Date Opened'] = temp['Date Opened'].dt.strftime('%m/%d/%Y')
temp.head()
```

```python
temp['Date Opened'] = temp['Date Opened'].astype('datetime64[ns]')
temp['Year'] = temp['Date Opened'].dt.strftime('%Y')
temp.head()
```

```python
temp['Days Open'] = temp['Days Open'].astype('float64')
temp['Year'] = temp['Year'].astype('object')
temp['Change Control'] = temp['Change Control'].astype('int64')
temp.to_csv('temp.csv', index=False)
```

```python
Total_Year = temp.groupby('Year').agg({ "Days Open": "mean"})
df = pd.DataFrame(Total_Year)
df['Days Open'] = df['Average Days Open'](df['Average Days Open'])
df.to_csv('Year.csv', index=True)
df.head()
```
